```python
%matplotlib inline
import yt
import trident

import seaborn as sns
```

```python
def load_ds(model, ion_list = ['H I', 'O VI', 'N V']):
    # only works on Iryna's home computer for now
    if model == 'stream':
        base = 'g160_torB_s4'
    elif model == 'anisd':
        base = 'g160_torB_ad'
    ds = yt.load('~/Work/galaxy/%s/DD2600'%(base))
    trident.add_ion_fields(ds, ions=ion_list)
    return ds
```

```python
ds_stream = load_ds('stream')
ds_anisd  = load_ds('anisd' )
```

```python
ad_stream = ds_stream.all_data()
ad_anisd  = ds_anisd.all_data()
```

```python
def phase_details(p, xfield, yfield, zfield):
    p.set_log(xfield[1], False)
    p.set_log(yfield[1], True)
    p.set_unit(xfield, 'km/s')
    
```

```python
xfield = ('gas', 'velocity_magnitude')
yfield = ('gas', 'temperature')
zfield = ('gas', 'O_p5_number_density')
p_stream = yt.PhasePlot(ad_stream, xfield, yfield, zfield, weight_field = None)
```

```python
phase_details(p_stream, xfield, yfield, zfield)
p_stream.show()
```
