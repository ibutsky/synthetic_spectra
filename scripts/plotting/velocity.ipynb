```python
import os
import sys
import numpy as np
import glob

import matplotlib.pylab as plt
from matplotlib.colors import LogNorm
import ion_plot_definitions as ipd

import plotting_tools as pt 

sys.path.append('../analysis')
import spectrum_analysis_tools as spa

import seaborn as sns
import palettable
#sns.set_style("ticks",{'axes.grid': True, "ytick.major.size": 0.1,
#                "ytick.minor.size": 0.05,
#                'grid.linestyle': '--'
#            })
```

```python
pal = palettable.cmocean.sequential.Tempo_4_r.mpl_colors

ion_list = ['H I', 'Si II', 'Si III', 'O VI']
color_list = ['tan', 'purple', 'cornflowerblue', 'orange']
color_list = pal

sim_name= "P0_agncr"
data_loc = '../../data/analyzed_spectra'

sightline = 36
spec_path = '%s/COS-FUV_%s_z0.25_%i'%(data_loc, sim_name, sightline)
spec = os.path.basename(spec_path)

plot_name = '../../plots/velocity/%s.png'%(spec)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(3.8, 2.2))

for ion, color in zip(ion_list, color_list):
    vv, flux, vvfit, fluxfit,  wl, wlfit, w0 = spa.load_velocity_data(ion, spec)
    new_v = np.ravel(np.column_stack((vv, vv)))
    new_f = np.ravel(np.column_stack((flux, flux)))
    ax.plot(new_v[1:], new_f[:-1], color = 'black',linewidth = 0.5, alpha = 0.2, zorder = 0)
    if os.path.isfile('%s/FitInspection.fits'%(spec_path)):
        ax.plot(vvfit, fluxfit, color = 'black', linewidth = 4, alpha = 0.7)
        ax.plot(vvfit, fluxfit, color = color, label = ion, linewidth = 2, alpha = 0.8)

ax.set_xlim(-295, 295)
ax.set_ylim(-0.05, 1.3)
ax.set_xlabel('Relative Velocity (km/s)')
ax.set_ylabel('Normalized Flux')
#ax.legend(frameon = False, fontsize = 'small')

fig.tight_layout()

#plt.show()
plt.savefig("vel_cr.png", dpi = 300, transparent = True)

```
